```python
with open('corpus/Tagged/porcodioaugu.xml') as fd:
#with open('1897LENZ-11-TAGGED.xml') as fd:
    doc = xmltodict.parse(fd.read())
for sentence in doc['TEI']['text']['body']['p']:
    sentence_word = []
    sentence_lemma = []
    if sentence['@xml:lang'] == 'arn':
        # sentence with length > 1
        if (type(sentence['w']) is list):
            for word in sentence['w']:
                #lemmas.append(word['@lemma'])
                sentence_lemma.append(word['@lemma'])
                
                # for case
                # text in <w>
                # <w xml:lang="arn" lemma="paliwe" pos="N" corresp="palin pitch">paliwe</w>
                # text in <w> is '-'
                # <w xml:lang="arn" lemma="ketrel-piriña" pos="N" corresp="chin break">
                #     <m baseForm="ketre" type="root" corresp="chin">ketre</m>
                #     <m baseForm="el" type="nmlz">l</m>
                #     -
                #     <m baseForm="piriña" type="root" corresp="break">piriña</m>
                # </w>
                if ('#text' in word.keys() and word['#text'] != '-'):
                    #words.append(word['#text'])
                    sentence_word.append(word['#text'])
                else:
                    
                    # word has different subwords
                    if (type(word['m']) is list):
                        string = ""
                        for subs in word['m']:
                            if (type(subs) is str):
                                string = string + subs
                            else:
                                #print (subs)
                                #print(subs.keys())
                                if ('#text' in subs.keys()):
                                    string = string + subs['#text']
                        #words.append(string)
                        sentence_word.append(string)
                    else:
                        # for case
                        # <w xml:lang="arn" lemma="meñcholongko" pos="NP" corresp="Meñcholongko"><m>Meñcholongko</m></w>
                        if (type(word['m']) is str):
                            #words.append(word['m'])
                            sentence_word.append(word['m'])
                        else:
                            #words.append(word['m']['#text'])
                            sentence_word.append(word['m']['#text'])
        
        # sentence with length = 1
        else:
            #lemmas.append(sentence['w']['@lemma'])
            sentence_lemma.append(sentence['w']['@lemma'])
            if (type(sentence['w']['m']) is list):
                string = ""
                for subs in sentence['w']['m']:
                    if (type(subs) is str):
                        string = string + subs
                    else:
                        # for case there is a subword but no morphe
                        # <m baseForm="iy" type="ind3"></m>
                        if ('#text' in subs.keys()):
                            string = string + subs['#text']
                #words.append(string)
                sentence_word.append(string)
            else:
                if (type(sentence['w']['m']) is str):
                    #words.append(sentence['w']['m'])
                    sentence_word.append(sentence['w']['m'])
                else:
                    #words.append(sentence['w']['m']['#text'])
                    sentence_word.append(sentence['w']['m']['#text'])
    lemmas.append(sentence_lemma)
    words.append(sentence_word)
```

```python
k = 0
n = 4000



# extract from xml
words_context = []
lemmas_context = []
for idx_s in range(len(words)):
    for idx_w in range(len(words[idx_s])):
        lemmas_context.append(" ".join(lemmas[idx_s][idx_w]))
        leftcount = 0
        rightcount = 0
        leftidx = idx_w
        rightidx = idx_w
        
        words_context.append(LC + " " + " ".join(words[idx_s][idx_w]) + " " + RC)
        
        # while not out of bound from sublist
        while ((leftidx-1) >= 0 and leftcount < k):
            flag = False
            # move pointer up
            leftidx = leftidx - 1
            leftword = words[idx_s][leftidx]
            for wordidx in reversed(range(len(leftword))):
                if (leftcount < k):
                    words_context[-1] = leftword[wordidx] + " " + words_context[-1]
                    leftcount = leftcount + 1
                # reach wordcount in current word
                else:
                    flag = True
                    break
            if (flag == False and leftcount<k):
                words_context[-1] = S + " " + words_context[-1]
                leftcount = leftcount + 1

        while ((rightidx+1) <= (len(words[idx_s]) - 1) and rightcount < k):
            flag = False
            # move pointer down
            rightidx = rightidx + 1
#             print(idx_s)
#             print(rightidx)
#             print(len(words))
#             print(len(words[idx_s]))
#             print(" ")
            rightword = words[idx_s][rightidx]
            for wordidx in range(len(rightword)):
                if (rightcount< k):
                    words_context[-1] = words_context[-1] + " " + rightword[wordidx]
                    rightcount = rightcount + 1
                else:
                    flag = True
                    break
            if (flag == False and rightcount<k):
                words_context[-1] = words_context[-1] + " " + S
                rightcount = rightcount + 1
```

```python
change=("ə","ü")
def process_spelling(pool,change):
    pool_normalized=[]
    source=change[0]
    target=change[1]
    for s in pool:
        if source in s:
            s=s.replace(source,target)
            pool_normalized.append(s)
    return pool_normalized

#make pools of tagged data

flat_words = [item for sublist in words for item in sublist]
flat_lemmas= [item for sublist in lemmas for item in sublist]

porcodiaz=process_spelling(flat_words,change)

count=0
for i in porcodiaz:
    if "ə" in i:
        count+=1
print(count)
```
